```python
import numpy as np
import matplotlib.pyplot as plt

# definisi vektor
v = np.array([2,3])
s = np.array([1,0])

print("Vektor v:", v)
print("Vektor s:", s)

# hasil penjumlahan
hasil = v + s
print("Hasil v + s:", hasil)

# gabungkan semua vektor
vecs = np.array([v, s, hasil])

# origin untuk tiap vektor
origins = np.zeros((3,2))

# plot sumbu
plt.axhline(0)
plt.axvline(0)

# gambar vektor
plt.quiver(origins[:,0], origins[:,1],
           vecs[:,0], vecs[:,1],
           angles='xy', scale_units='xy', scale=1)

plt.xlim(-5,5)
plt.ylim(-5,5)
plt.gca().set_aspect('equal')

plt.grid()
plt.title('Visualisasi v, s, dan v + s')

plt.show()
```
